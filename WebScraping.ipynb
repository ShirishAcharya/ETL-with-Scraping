```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://findajob.dwp.gov.uk/search?q=data+engineer&w=UK"

response = requests.get(url)

response
```

```python
soup = BeautifulSoup(response.content,"html.parser")
```

```python
soup
```

```python
jobs_listings = soup.find_all("div", class_="search-result")
```

```python
jobs_listings
```

```python
titles = []
links = []
for job in jobs_listings:
    title_element = job.find("a", class_="govuk-link")
    if title_element:
        titles.append(title_element.text.strip()) 
        links.append(title_element["href"])
```

```python
titles
```

```python
links
```

```python
dates = []

for job in jobs_listings:
    date_element = job.find("li")
    if date_element:
        dates.append(date_element.text.strip()) 
```

```python
dates
```

```python
companies = []

for job in jobs_listings:
    company_element = job.find("strong")
    if company_element:
        companies.append(company_element.text.strip())
```

```python
companies
```

```python
location = []

for job in jobs_listings:
    location_element = job.find("span")
    if location_element:
        location.append(location_element.text.strip())
```

```python
location
```

```python
salary = []

for job in jobs_listings:
    salary_element = job.find("li", string = lambda text: text and "£" in text)
    if salary_element:
        salary.append(salary_element.text.strip())
    else:
        salary.append(None)
```

```python
salary
```

```python
df = pd.DataFrame({
    "title" : titles,
    "links" : links,
    "date_posted" : dates,
    "company" : companies,
    "location" : location,
    "salary" : salary,
}
)
```

```python
df
```

### Transformations

```python
df["date_posted"] = pd.to_datetime(df["date_posted"])
```

```python
df
```

```python
def experience_level(title):
    title_lower = title.lower()
    if "senior" in title_lower:
        return "Senior"
    elif "lead" in title_lower:
        return "Lead"
    elif "head" in title_lower:
        return "Head"
    elif "junior" in title_lower:
        return "Junior"
    else:
        return "Entry"
    
```

```python
df["experience_level"] = df["title"].apply(experience_level) 
```

```python
df
```

```python
df.to_csv("jobs_listings.csv")
```
